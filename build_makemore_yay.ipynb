```python
import torch
```

```python
TOKEN_START = '<S>'
TOKEN_END = '<E>'

def char_to_index(c):
    if c == TOKEN_START:
        return 0
    if c == TOKEN_END:
        return 27
    return ord(c) - ord('a') + 1

def index_to_char(i):
    if i == 0:
        return TOKEN_START
    if i == 27:
        return TOKEN_END
    return chr(i + ord('a') - 1)

def get_bigrams(words):
    bigrams = torch.zeros((28, 28))

    for w in words:
        cs = [TOKEN_START] + list(w) + [TOKEN_END]
        for c1, c2 in zip(cs, cs[1:]):
            i1 = char_to_index(c1)
            i2 = char_to_index(c2)
            bigrams[i1, i2] += 1

    return bigrams
```

```python
names = list(map(lambda line: line.strip(), open('data/names.txt', 'r').readlines()))
bigrams = get_bigrams(names)
```
